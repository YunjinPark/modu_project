단어 세트 4개 만들어 하나의 파일에 저장. 

```python
import json
```

```python
filename = 'big.json'
with open(filename, 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
print('데이터 분류')
for i in data.keys():
    if i: print("'"+i+"',", end="")
```

```python
g1 = ['gen','ind','art_ind','art']
g2 = ['다큐멘터리','멜로로맨스','서부극(웨스턴)','범죄','뮤지컬',
      '가족','액션','기타','공연','미스터리','성인물(에로)','스릴러',
      '전쟁','코미디','어드벤처','공포(호러)','SF','드라마','판타지',
      '애니메이션','사극',]
```

```python
def make_txt(title, list_, num):
    '''리스트에서 num 개수만큼 뽑고 title도 붙여 string 형식으로 반환'''
    if len(list_) < num: 
        print(title, ' 단어 개수 부족함.')
        return False
    t = title + ': '
    for i in list_[:num-1]:
        t += i+', '
    t +=list_[num-1]
    return t
```

```python
def save_file(file_path, title, t1, t2, t3, t4):
    '''weat를 계산하기 위해 한 t1, t2, t3, t4 문장을 한 파일에 저장'''
    with open(file_path + title, 'w') as f:
        f.write(t1+'\n')
    with open(file_path + title, 'a') as f:
        f.write(t2+'\n')    
        f.write(t3+'\n')
        f.write(t4+'\n')
```

```python
print('일반영화 단어')
print(make_txt(g1[0], data[g1[0]],0))
```

```python
print('예술영화 단어')
print(make_txt(g1[3], data[g1[3]],20))
```

```python
print('다큐멘터리 단어')
print(make_txt(g2[0], data[g2[0]],20))
```

```python
text = []
for i in g1:
    text.append(make_txt(i, data[i], 30))
```

```python
print('make_txt 결과')
text
```

```python
text2 = []
for i in g2:
    text2.append(make_txt(i, data[i], 30))
```

```python
print('make_txt 결과')
text2
```

```python
file_path = "./weat_keywords/gen_art_2/"
for i in range(len(text2)):
    for j in range(len(text2)):
        save_file(file_path, 'gen_'+text[3].split(':')[0]+'_'+text2[i].split(':')[0]+'_'+text2[j].split(':')[0], 
                      text[0], text[3], text2[i], text2[j])
```

```python
for k in range(1, 4):
    file_path = "./weat_keywords/gen_" + text[k].split(':')[0]+'/'
    for i in range(len(text2)):
        for j in range(i+1, len(text2)):
            save_file(file_path, 'gen_'+text[k].split(':')[0]+'_'+text2[i].split(':')[0]+'_'+text2[j].split(':')[0], 
                      text[0], text[k], text2[i], text2[j])
```
